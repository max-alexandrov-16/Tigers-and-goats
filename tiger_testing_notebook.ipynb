```python
from run_game import run_episode
from tiger_agent import TigerAgent, Model
```

```python
board_size = 5
model = Model(board_size)
reward_scheme = {"eating": 2, "winning": 3, "losing": -3,"no score":-1}
tiger_agent = TigerAgent(model,board_size,reward_scheme,None)
tiger_agent.load_model('model_v3_size_5.pth')
number_of_test_runs = 500
tiger_wins_count = 0
tiger_losses_count = 0
for run in range(number_of_test_runs):
    _,_,_,tiger_rewards = run_episode(board_size,30,reward_scheme,tiger_agent=tiger_agent,verbose = False)
    print(tiger_rewards)
    if tiger_rewards[-1] == reward_scheme["winning"]:
        tiger_wins_count += 1
    elif tiger_rewards[-1] == reward_scheme["losing"]:
        tiger_losses_count += 1
print(f'tiger won {tiger_wins_count} times out of {number_of_test_runs} runs, which accounts for {tiger_wins_count/number_of_test_runs*100}% of the runs')
print(f'tiger lost {tiger_losses_count} times out of {number_of_test_runs} runs, which accounts for {tiger_losses_count/number_of_test_runs*100}% of the runs')
```

```python
random_tiger_wins_count = 0
for run in range(number_of_test_runs):
    _,_,_,tiger_rewards = run_episode(board_size,40,tiger_agent=None,verbose = False)
    if tiger_rewards[-1] == 5:
        random_tiger_wins_count += 1

print(f'tiger won {random_tiger_wins_count} times out of {number_of_test_runs} runs, which accounts for {random_tiger_wins_count/number_of_test_runs*100}% of the runs')
```
